<H5>1.3. Medidas Estatísticas Descritivas do Conjunto de Dados - Medidas de Frequência, Medidas de Tendência Central e Medidas de Dispersão:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_cinema.csv')

#Calcular as Medidas Estatísticas Descritivas para todas as Variáveis do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados = conjunto_dados.describe(include='all')

#Selecionar as Variáveis Numéricas do Conjunto de Dados:
variáveis_numéricas_conjunto_dados = conjunto_dados.select_dtypes(include=['number'])

#Calcular a Moda das Variáveis Numéricas do Conjunto de Dados:
moda_variáveis_numéricas_conjunto_dados = variáveis_numéricas_conjunto_dados.mode().iloc[0]

#Adicionar a Moda às Medidas Estatísticas Descritivas das Variáveis Numéricas do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados.loc['Moda'] = moda_variáveis_numéricas_conjunto_dados

#Calcular a Amplitude das Variáveis Numéricas do Conjunto de Dados:
amplitude_variáveis_numéricas_conjunto_dados = variáveis_numéricas_conjunto_dados.max() - variáveis_numéricas_conjunto_dados.min()

#Adicionar a Amplitude às Medidas Estatísticas Descritivas das Variáveis Numéricas do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados.loc['Amplitude'] = amplitude_variáveis_numéricas_conjunto_dados

#Calcular a Variância das Variáveis Numéricas do Conjunto de Dados:
variância_variáveis_numéricas_conjunto_dados = variáveis_numéricas_conjunto_dados.var()

#Adicionar a Variância às Medidas Estatísticas Descritivas das Variáveis Numéricas do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados.loc['Variância'] = variância_variáveis_numéricas_conjunto_dados

#Traduzir os Índices das Medidas Estatísticas Descritivas de Inglês para Português:
medidas_estatísticas_descritivas_variáveis_conjunto_dados.rename(index={
    'count': 'Contagem',
    'mean': 'Média',
    'std': 'Desvio Padrão',
    'min': 'Mínimo',
    '25%': '1.º Quartil',
    '50%': 'Mediana',
    '75%': '3.º Quartil',
    'max': 'Máximo',
    'Amplitude': 'Amplitude',
    'Variância': 'Variância'}, inplace=True)

#Ordenar as Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados_ordenadas = [
    'Contagem', 'Média', 'Mediana', 'Moda', 'Desvio Padrão', 'Mínimo', 
    '1.º Quartil', '3.º Quartil', 'Máximo', 'Amplitude', 'Variância']

#Reordenar as Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados segundo a Ordem Definida:
medidas_estatísticas_descritivas_variáveis_conjunto_dados = medidas_estatísticas_descritivas_variáveis_conjunto_dados.loc[medidas_estatísticas_descritivas_variáveis_conjunto_dados_ordenadas]

#Arredondar os Resultados para 2 Casas Decimais:
medidas_estatísticas_descritivas_variáveis_conjunto_dados = medidas_estatísticas_descritivas_variáveis_conjunto_dados.round(2)

#Configurações da Exibição de Resultados - Largura Máxima de Exibição do DataFrame:
pd.set_option('display.width', 1000)

#Configurações da Exibição de Resultados - Mostrar Todas as Variáveis do Conjunto de Dados numa única Linha:
pd.set_option('display.max_columns', None)

#Mostrar as Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados:
print(medidas_estatísticas_descritivas_variáveis_conjunto_dados.to_string(index=True))
```

<H5>1.5.1. Histogramas para as Variáveis Numéricas com as Distribuições Normal, Exponencial e Log-Normal Ajustadas:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar o Submódulo Pyplot da Biblioteca Matplotlib:
import matplotlib.pyplot as plt

#A partir do Submódulo Pyplot da Biblioteca Matplotlib, Importar as Funções Mostrar e Subgráficos:
from matplotlib.pyplot import show, subplots

#A partir do Submódulo Ticker da Biblioteca Matplotlib, Importar a Função Formatador de Função:
from matplotlib.ticker import FuncFormatter

#Importar a Biblioteca Numerical Python:
import numpy as np

#A partir do Submódulo Stats da Biblioteca SciPy, Importar as Distribuições Estatísticas Normal, Exponencial e Log-Normal:
from scipy.stats import norm, expon, lognorm

#Definir o Tipo de Letra dos Textos (sans-serif):
plt.rcParams['font.family'] = 'sans-serif'

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_cinema.csv')

#Obter as Variáveis Numéricas do Conjunto de Dados:
lista_variáveis_numéricas = conjunto_dados.select_dtypes(include=['number']).columns.tolist()

#Definir o Número de Colunas:
colunas = 4

#Definir a Altura e a Largura dos Histogramas:
altura_largura_histogramas = 8

#Verificar Se Existem Variáveis Numéricas:
if lista_variáveis_numéricas:

    #Calcular o Número Total de Variáveis Numéricas:
    número_total_variáveis_numéricas = len(lista_variáveis_numéricas)
    
    #Calcular o Número de Linhas:
    número_linhas = (número_total_variáveis_numéricas // colunas) + (número_total_variáveis_numéricas % colunas > 0)

    #Criar a Figura e a Matriz dos Histogramas:
    figura_histogramas, matriz_histogramas = subplots(número_linhas, colunas, figsize=(colunas * altura_largura_histogramas, número_linhas * altura_largura_histogramas), squeeze=False)

    #Configurações da Figura - Título Geral - Designação, Cor e Tamanho:
    figura_histogramas.suptitle("Conjunto de Dados 1 - Histogramas das Variáveis Numéricas com as Distribuições Normal, Exponencial e Log-Normal Ajustadas", 
                                color='black', fontsize=25, x=0.5)

    #Configurações da Figura - Margens Superior, Direita, Inferior e Esquerda:
    figura_histogramas.subplots_adjust(top=0.92, right=0.95, bottom=0.35, left=0.15)

    #Configurações da Figura - Espaçamentos Horizontal e Vertical:
    figura_histogramas.subplots_adjust(wspace=0.3, hspace=0.4)

    #Iniciar as Coordenadas X e Y para Navegação na Matriz dos Histogramas:
    X, Y = 0, 0

    #Iterar sobre o Índice das Variáveis Numéricas:
    for índice_variável_numérica in range(número_total_variáveis_numéricas):

        #Obter os Dados da Variável Numérica Atual:
        dados_variável_numérica_atual = conjunto_dados[lista_variáveis_numéricas[índice_variável_numérica]].dropna().values

        #Criar os Histogramas para as Variáveis Numéricas - Configurações das Barras - Cores e Número de Intervalos:
        matriz_histogramas[X, Y].hist(dados_variável_numérica_atual, color='#2ca02c', edgecolor='#2ca02c', bins='auto', density=True)

        #Configurações dos Histogramas - Título - Designação, Cor e Tamanho:
        matriz_histogramas[X, Y].set_title(f"Histograma para a Variável {lista_variáveis_numéricas[índice_variável_numérica]}\n", color='black', fontsize=19)

        #Configurações do Eixo Atual - Definir Exibição dos Elementos Abaixo do Eixo Atual:
        matriz_histogramas[X, Y].set_axisbelow(True)

        #Configurações do Eixo X - Título - Designação, Cor e Tamanho:
        matriz_histogramas[X, Y].set_xlabel(f"Variável Numérica {lista_variáveis_numéricas[índice_variável_numérica]}", color='black', fontsize=19)

        #Configurações do Eixo X - Rótulos dos Marcadores - Cor e Tamanho:
        matriz_histogramas[X, Y].tick_params(axis='x', colors='black', labelsize=16)

        #Configurações do Eixo Y - Título - Designação, Cor e Tamanho:
        matriz_histogramas[X, Y].set_ylabel("Densidade de Probabilidade", color='black', fontsize=19)

        #Configurações do Eixo Y - Rótulos dos Marcadores - Cor e Tamanho:
        matriz_histogramas[X, Y].tick_params(axis='y', colors='black', labelsize=16)

        #Configurações do Eixo Y - Valores dos Rótulos - Formatação:
        matriz_histogramas[X, Y].yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.3f}'.replace(",", ".")))

        #Configurações do Eixo Y - Linhas de Grelha - Cor, Estilo de Linha e Transparência:
        matriz_histogramas[X, Y].grid(True, axis='y', color='gray', linestyle='-', alpha=0.3)

        #Configurações das Margens dos Eixos X e Y - Cor e Espessura:
        for margem in matriz_histogramas[X, Y].spines.values():
            margem.set_color('black')
            margem.set_linewidth(2.5)
        
        #Ocultar as Margens Superior e Direita:
        matriz_histogramas[X, Y].spines['top'].set_visible(False)
        matriz_histogramas[X, Y].spines['right'].set_visible(False)

        #Gerar Pontos para Ajustar as Distribuições aos Histogramas:
        pontos_ajustar_distribuições = np.linspace(pd.Series(dados_variável_numérica_atual).quantile(0.001), pd.Series(dados_variável_numérica_atual).quantile(0.999), 100)

        #Calcular a Média e o Desvio Padrão para Ajustar a Distribuição Normal aos Dados dos Histogramas:
        média, desvio_padrão = norm.fit(dados_variável_numérica_atual)

        #Ajustar a Distribuição Normal aos Dados dos Histogramas - Configurações da Distribuição Normal - Designação e Cor:
        matriz_histogramas[X, Y].plot(pontos_ajustar_distribuições, norm.pdf(pontos_ajustar_distribuições, média, desvio_padrão), label="Distribuição Normal", color='#ff7f0e')

        #Calcular a Localização e a Escala para Ajustar a Distribuição Exponencial aos Dados dos Histogramas:
        localização, escala = expon.fit(dados_variável_numérica_atual)

        #Ajustar a Distribuição Exponencial aos Dados dos Histogramas - Configurações da Distribuição Exponencial - Designação e Cor:
        matriz_histogramas[X, Y].plot(pontos_ajustar_distribuições, expon.pdf(pontos_ajustar_distribuições, localização, escala), label="Distribuição Exponencial", color='#87CEEB')

        #Filtrar os Dados Positivos da Variável Numérica Atual:
        dados_positivos_variável_numérica_atual = dados_variável_numérica_atual[dados_variável_numérica_atual > 0]
        
        #Verificar Se a Variável Numérica Atual Apresenta Dados Positivos:
        if len(dados_positivos_variável_numérica_atual) > 0:
            
            #Calcular a Forma, a Localização e a Escala para Ajustar a Distribuição Log-Normal aos Dados dos Histogramas:
            forma, localização, escala = lognorm.fit(dados_positivos_variável_numérica_atual, floc=0)
            
            #Ajustar a Distribuição Log-Normal aos Dados dos Histogramas - Configurações da Distribuição Log-Normal - Designação e Cor:
            matriz_histogramas[X, Y].plot(pontos_ajustar_distribuições, lognorm.pdf(pontos_ajustar_distribuições, forma, localização, escala), label="Distribuição Log-Normal", color='#9467bd')
    
        #Senão: Não Existem Dados para Ajustar a Distribuição Log-Normal aos Dados dos Histogramas:
        else:
            matriz_histogramas[X, Y].text(0.5, 0.5, 'Não Existem Dados para Ajustar a Distribuição Log-Normal', horizontalalignment='center', verticalalignment='center', color='black', fontsize=12, transform=matriz_histogramas[X, Y].transAxes)

        #Mostrar a Legenda para a Distribuição Ajustada aos Dados dos Histogramas:
        matriz_histogramas[X, Y].legend(fontsize=10, loc='best')

        #Atualizar as Coordenadas X e Y:
        if Y == colunas - 1:
            X += 1
            Y = 0
        else:
            Y += 1

    #Remover os Subgráficos Vazios:
    total_subgráficos = número_linhas * colunas
    if número_total_variáveis_numéricas < total_subgráficos:
        for índice in range(número_total_variáveis_numéricas, total_subgráficos):
            figura_histogramas.delaxes(matriz_histogramas.flatten()[índice])

#Mostrar a Figura com os Histogramas e as Distribuições Ajustadas:
show()
```

<H5>1.5.2. Gráfico de Barras para a Componente Temporal Ano da Variável Temporal Data:</H5>

```python
#Importar o Pandas:
import pandas as pd

#Importar o Submódulo Pyplot da Biblioteca Matplotlib:
import matplotlib.pyplot as plt

#A partir do Submódulo Ticker da Biblioteca Matplotlib, Importar a Função Formatador de Função:
from matplotlib.ticker import FuncFormatter

#A partir do Submódulo Figure da Biblioteca Matplotlib, Importar a Classe Figura:
from matplotlib.figure import Figure

#Importar a Biblioteca Numpy:
import numpy as np

#A partir do Módulo Funções do DSLabs, Importar as Funções Obter Tipos de Variáveis e Altura:
from dslabs_functions import get_variable_types, HEIGHT

#Ler e Carregar o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_cinema.csv')

#Definir Função para Derivar os Componentes Temporais a partir da Variável Temporal Data:
def função_derivar_componentes_temporais(dados: pd.DataFrame, variáveis_temporais: list[str]) -> pd.DataFrame:
    
    #Iterar sobre as Variáveis Temporais:
    for variável_temporal in variáveis_temporais:
        
        #Converter a Variável Temporal para o Formato Datetime:
        dados[variável_temporal] = pd.to_datetime(dados[variável_temporal])
        
        #Derivar a Componente Temporal Ano a partir da Variável Temporal Data:
        dados[variável_temporal + "_Ano"] = dados[variável_temporal].dt.year
    
    #Devolver os Dados com os Componentes Temporais Derivados:
    return dados

#Definir Função para Criar os Gráficos de Barras e Analisar a Granularidade Temporal:
def função_criar_gráficos_barras_analisar_granularidade_temporal(dados: pd.DataFrame, variável: str) -> None:
    
    #Definir os Níveis de Detalhe Temporais (Ano):
    níveis_detalhe = ["Ano"]
    
    #Calcular o Número de Colunas:
    número_colunas: int = len(níveis_detalhe)
    
    #Criar a Figura e os Eixos para os Subgráficos:
    figura, eixos = plt.subplots(1, número_colunas, figsize=(número_colunas * HEIGHT, HEIGHT))
    
    #Se Existir Apenas um Eixo:
    if número_colunas == 1:
        
        #Transformar Eixo numa Lista para Permitir Indexação:
        eixos = [eixos]
    
    #Configurações dos Gráficos de Barras - Título Geral - Designação Cor e Tamanho:
    figura.suptitle("Conjunto de Dados 1 - Gráfico de Barras para a Componente Temporal Ano Derivada da Variável Temporal Data\n", color='black', fontsize=10)
    
    #Iterar sobre os Índices dos Níveis de Detalhe:
    for índice_variável_temporal, detalhe in enumerate(níveis_detalhe):
        
        #Obter e Ordenar as Contagens de cada Valor Único da Componente Temporal:
        contagens = dados[f"{variável}_{detalhe}"].value_counts().sort_index()
        
        #Criar Lista de Rótulos de Texto para os Valores Únicos:
        rótulos_variáveis_temporais = contagens.index.astype(str).to_list()
        
        #Criar o Gráfico de Barras:
        eixos[índice_variável_temporal].bar(rótulos_variáveis_temporais, contagens.values, color='#2ca02c', edgecolor='#2ca02c')
        
        #Configurações do Eixo X:
        eixos[índice_variável_temporal].set_xlabel(f"\nComponente Temporal {detalhe}", color='black', fontsize=10)
        eixos[índice_variável_temporal].tick_params(axis='x', colors='black')
        
        #Configurações do Eixo Y:
        eixos[índice_variável_temporal].set_ylabel("Número de Registos\n", color='black', fontsize=10)
        eixos[índice_variável_temporal].tick_params(axis='y', colors='black')
        eixos[índice_variável_temporal].set_ylim(0, 160000)
        
        #Configurações do Eixo Y - Valores - Formatação:
        eixos[índice_variável_temporal].yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"{x:,.0f}".replace(',', '.')))
        
        #Configurações das Margens dos Eixos - Cor e Espessura:
        for margem in ['bottom', 'left', 'top', 'right']:
            eixos[índice_variável_temporal].spines[margem].set_color('black')
            eixos[índice_variável_temporal].spines[margem].set_linewidth(2.5)
    
    #Ajustar o Layout para Evitar Sobreposição:
    figura.tight_layout()
    
    #Mostrar o Gráfico de Barras:
    plt.show()

#Obter os Tipos de Variáveis:
tipos_variáveis: dict[str, list] = get_variable_types(conjunto_dados)

#Derivar as Componentes Temporais a partir da Variável Temporal Data:
componentes_temporais: pd.DataFrame = função_derivar_componentes_temporais(conjunto_dados, tipos_variáveis["date"])

#Iterar sobre as Variáveis Temporais para Análise:
for variável_temporal in tipos_variáveis["date"]:
    
    #Chamar a Função para Criar Gráficos de Barras e Analisar a Granularidade Temporal:
    função_criar_gráficos_barras_analisar_granularidade_temporal(componentes_temporais, variável_temporal)
```

<H5>1.5.4. Estudo dos Valores Atípicos:</H5>

<H5>1.5.4.1. Estudo dos Valores Atípicos - Resultados:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar a Biblioteca Numerical Python:
import numpy as np

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_cinema.csv')

#Definir o Fator do Intervalo Interquartil:
fator_intervalo_interquartil = 1.5

#Definir o Fator do Desvio Padrão:
fator_desvio_padrão = 3

#Verificar Se Existem Variáveis Numéricas no Conjunto de Dados:
variáveis_numéricas = conjunto_dados.select_dtypes(include=[np.number]).columns

#Se Existirem Variáveis Numéricas no Conjunto de Dados:
if len(variáveis_numéricas) > 0:

    #Criar um Dicionário para Armazenar as Percentagens dos Valores Atípicos:
    dicionário_percentagens_valores_atípicos = {'Intervalo Interquartil': [], 'Desvio Padrão': []}

    #Iterar sobre as Variáveis Numéricas do Conjunto de Dados:
    for variável_numérica in variáveis_numéricas:
    
        #Calcular o 1.º Quartil:
        Q1 = conjunto_dados[variável_numérica].quantile(0.25)
        
        #Calcular o 3.º Quartil:
        Q3 = conjunto_dados[variável_numérica].quantile(0.75)
        
        #Calcular o Intervalo Interquartil:
        intervalo_interquartil = Q3 - Q1
        
        #Calcular o Limite Inferior - Intervalo Interquartil:
        limite_inferior_intervalo_interquartil = Q1 - fator_intervalo_interquartil * intervalo_interquartil

        #Calcular o Limite Superior - Intervalo Interquartil:
        limite_superior_intervalo_interquartil = Q3 + fator_intervalo_interquartil * intervalo_interquartil
    
        #Calcular a Percentagem de Valores Atípicos para cada Variável Numérica - Intervalo Interquartil:
        percentagem_valores_atípicos_intervalo_interquartil = ((conjunto_dados[variável_numérica] < limite_inferior_intervalo_interquartil) | 
                                                               (conjunto_dados[variável_numérica] > limite_superior_intervalo_interquartil)).mean() * 100
    
        #Adicionar as Percentagens de Valores Atípicos Obtidas ao Dicionário de Percentagens de Valores Atípicos - Intervalo Interquartil:
        dicionário_percentagens_valores_atípicos['Intervalo Interquartil'].append(f"{int(percentagem_valores_atípicos_intervalo_interquartil)}%")

        #Calcular a Média:
        média = conjunto_dados[variável_numérica].mean()
        
        #Calcular o Desvio Padrão:
        desvio_padrão = conjunto_dados[variável_numérica].std()
    
        #Calcular o Limite Inferior - Desvio Padrão:
        limite_inferior_desvio_padrão = média - fator_desvio_padrão * desvio_padrão

        #Calcular o Limite Superior - Desvio Padrão:
        limite_superior_desvio_padrão = média + fator_desvio_padrão * desvio_padrão
    
        #Calcular a Percentagem de Valores Atípicos para cada Variável Numérica - Desvio Padrão:
        percentagem_valores_atípicos_desvio_padrão = ((conjunto_dados[variável_numérica] < limite_inferior_desvio_padrão) | 
                                                      (conjunto_dados[variável_numérica] > limite_superior_desvio_padrão)).mean() * 100
    
        #Adicionar as Percentagens de Valores Atípicos Obtidas ao Dicionário de Percentagens de Valores Atípicos - Desvio Padrão:
        dicionário_percentagens_valores_atípicos['Desvio Padrão'].append(f"{int(percentagem_valores_atípicos_desvio_padrão)}%")

    #Título Geral - Percentagem de Valores Atípicos para cada Variável Numérica - Intervalo Interquartil e Desvio Padrão:
    print("Percentagem de Valores Atípicos para cada Variável Numérica - Intervalo Interquartil e Desvio Padrão:\n")
    
    #Títulos - Variável Numérica ; Percentagem de Valores Atípicos - Intervalo Interquartil ; Percentagem de Valores Atípicos - Desvio Padrão:
    print(f"{'Variável Numérica:':<30}{'Percentagem de Valores Atípicos - Intervalo Interquartil:':<70}{'Percentagem de Valores Atípicos - Desvio Padrão:\n':<70}")
    
    #Iterar Sobre os Índices e Nomes das Variáveis Numéricas:
    for índice_variável_numérica, nome_variável_numérica in enumerate(variáveis_numéricas):
        
        #Atribuir a Percentagem de Valores Atípicos pelo Intervalo Interquartil à Variável Numérica Correspondente:
        percentagem_valores_atípicos_intervalo_interquartil = dicionário_percentagens_valores_atípicos['Intervalo Interquartil'][índice_variável_numérica]

        #Atribuir a Percentagem de Valores Atípicos pelo Desvio Padrão à Variável Numérica Correspondente:
        percentagem_valores_atípicos_desvio_padrão = dicionário_percentagens_valores_atípicos['Desvio Padrão'][índice_variável_numérica]

        #Nome da Variável Numérica ; Percentagem de Valores Atípicos - Intervalo Interquartil ; Percentagem de Valores Atípicos - Desvio Padrão:
        print(f"{nome_variável_numérica:<30}{percentagem_valores_atípicos_intervalo_interquartil:<70}{percentagem_valores_atípicos_desvio_padrão:<70}")

#Senão: Não Existem Variáveis Numéricas.
else:
    print("Não Existem Variáveis Numéricas.")
```

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_cinema.csv')

#Conversão da Variável Ordinal Temporal Data para o Formato DateTime:
conjunto_dados['Data'] = pd.to_datetime(conjunto_dados['Data'], errors='coerce', dayfirst=True)

#Extrair a Componente Temporal Ano a partir da Variável Temporal Data:
conjunto_dados['Ano'] = conjunto_dados['Data'].dt.year

#Remover a Variável Temporal Data:
conjunto_dados.drop(columns=['Data'], inplace=True)

#Mover a Variável Ano para a Posição 0 do Conjunto de Dados:
variáveis = ['Ano'] + [variável for variável in conjunto_dados.columns if variável != 'Ano']
conjunto_dados_codificado = conjunto_dados[variáveis]

#Guardar o Conjunto de Dados Codificado por Extração de Componentes Temporais:
conjunto_dados_codificado.to_csv('conjunto_dados_codificado.csv', index=False)
print("O Conjunto de Dados Codificado por Extração de Componentes Temporais Foi Guardado com Sucesso em conjunto_dados_codificado.csv")

#Mostrar os 5 Primeiros Registos do Conjunto de Dados Codificado por Extração de Componentes Temporais:
conjunto_dados_codificado.head()
```

<H5>2.1.2. Medidas Estatísticas Descritivas do Conjunto de Dados Pós-Codificação das Variáveis - Medidas de Frequência, Medidas de Tendência Central e Medidas de Dispersão:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_codificado.csv')

#Calcular as Medidas Estatísticas Descritivas para as Variáveis do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados = conjunto_dados.describe(include='all')

#Calcular a Moda das Variáveis do Conjunto de Dados:
moda_variáveis_conjunto_dados = conjunto_dados.mode().iloc[0]

#Adicionar a Moda às Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados.loc['Moda'] = moda_variáveis_conjunto_dados

#Calcular a Amplitude das Variáveis do Conjunto de Dados:
amplitude_variáveis_conjunto_dados = conjunto_dados.max() - conjunto_dados.min()

#Adicionar a Amplitude às Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados.loc['Amplitude'] = amplitude_variáveis_conjunto_dados

#Calcular a Variância das Variáveis do Conjunto de Dados:
variância_variáveis_conjunto_dados = conjunto_dados.var()

#Adicionar a Variância às Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados.loc['Variância'] = variância_variáveis_conjunto_dados

#Traduzir os Índices das Medidas Estatísticas Descritivas de Inglês para Português:
medidas_estatísticas_descritivas_variáveis_conjunto_dados.rename(index={
    'count': 'Contagem',
    'mean': 'Média',
    'std': 'Desvio Padrão',
    'min': 'Mínimo',
    '25%': '1.º Quartil',
    '50%': 'Mediana',
    '75%': '3.º Quartil',
    'max': 'Máximo',
    'Amplitude': 'Amplitude',
    'Variância': 'Variância'}, inplace=True)

#Ordenar as Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados:
medidas_estatísticas_descritivas_variáveis_conjunto_dados_ordenadas = [
    'Contagem', 'Média', 'Mediana', 'Moda', 'Desvio Padrão', 'Mínimo', 
    '1.º Quartil', '3.º Quartil', 'Máximo', 'Amplitude', 'Variância']

#Reordenar as Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados segundo a Ordem Definida:
medidas_estatísticas_descritivas_variáveis_conjunto_dados = medidas_estatísticas_descritivas_variáveis_conjunto_dados.loc[medidas_estatísticas_descritivas_variáveis_conjunto_dados_ordenadas]

#Arredondar os Resultados para 2 Casas Decimais:
medidas_estatísticas_descritivas_variáveis_conjunto_dados = medidas_estatísticas_descritivas_variáveis_conjunto_dados.round(2)

#Configurações da Exibição de Resultados - Largura Máxima de Exibição do DataFrame:
pd.set_option('display.width', 1000)

#Configurações da Exibição de Resultados - Mostrar Todas as Variáveis do Conjunto de Dados numa única Linha:
pd.set_option('display.max_columns', None)

#Mostrar as Medidas Estatísticas Descritivas das Variáveis do Conjunto de Dados:
print(medidas_estatísticas_descritivas_variáveis_conjunto_dados)
```

<H5>2.3. Tratamento dos Valores Atípicos:</H5>

<H5>2.3.1. Tratamento dos Valores Atípicos utilizando o Truncamento aos Limites Inferior e Superior pelo método do Intervalo Interquartil:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_sem_valores_omissos.csv')

#Definir o Fator do Intervalo Interquartil:
fator_intervalo_interquartil = 1.5

#Iniciar o Contador do Número Total de Valores Atípicos do Conjunto de Dados:
contador_número_total_valores_atípicos = 0

#Iniciar o Contador do Número Total de Valores Atípicos do Conjunto de Dados Pós-Truncamento aos Limites Inferior e Superior:
contador_número_total_valores_atípicos_pós_truncamento = 0

#Dicionário para Armazenar a Percentagem de Valores Atípicos por Variável:
percentagem_valores_atípicos_variável = {}

#Dicionário para Armazenar a Percentagem de Valores Atípicos por Variável Pós-Truncamento aos Limites Inferior e Superior:
percentagem_valores_atípicos_variável_pós_truncamento = {}

#Iterar sobre as Variáveis do Conjunto de Dados:
for variável in conjunto_dados.columns:
    
    #Calcular o 1.º Quartil:
    Q1 = conjunto_dados[variável].quantile(0.25)
    
    #Calcular o 3.º Quartil:
    Q3 = conjunto_dados[variável].quantile(0.75)
    
    #Calcular o Intervalo Interquartil:
    intervalo_interquartil = Q3 - Q1
    
    #Calcular o Limite Inferior - Intervalo Interquartil:
    limite_inferior = Q1 - fator_intervalo_interquartil * intervalo_interquartil

    #Calcular o Limite Superior - Intervalo Interquartil:
    limite_superior = Q3 + fator_intervalo_interquartil * intervalo_interquartil

    #Contar o Número Total de Valores Atípicos para cada Variável:
    número_total_valores_atípicos_variável = ((conjunto_dados[variável] < limite_inferior) | (conjunto_dados[variável] > limite_superior)).sum()

    #Contar o Número Total de Valores Atípicos do Conjunto de Dados Pré-Truncamento:
    contador_número_total_valores_atípicos += número_total_valores_atípicos_variável

    #Calcular a Percentagem de Valores Atípicos Pré-Truncamento:
    percentagem_valores_atípicos_variável[variável] = (número_total_valores_atípicos_variável / len(conjunto_dados)) * 100

    #Se Existirem Valores Atípicos na Variável:
    if número_total_valores_atípicos_variável > 0:

        #Aplicar o Truncamento para os Valores Atípicos:
        conjunto_dados[variável] = conjunto_dados[variável].clip(lower=limite_inferior, upper=limite_superior)

    #Contar o Número Total de Valores Atípicos para cada Variável Pós-Truncamento aos Limites Inferior e Superior:
    número_total_valores_atípicos_variável_pós_truncamento = ((conjunto_dados[variável] < limite_inferior) | (conjunto_dados[variável] > limite_superior)).sum()
    
    #Contar o Número Total de Valores Atípicos do Conjunto de Dados Pós-Truncamento aos Limites Inferior e Superior:
    contador_número_total_valores_atípicos_pós_truncamento += número_total_valores_atípicos_variável_pós_truncamento

    #Calcular a Percentagem de Valores Atípicos Pós-Truncamento aos Limites Inferior e Superior:
    percentagem_valores_atípicos_variável_pós_truncamento[variável] = (número_total_valores_atípicos_variável_pós_truncamento / len(conjunto_dados)) * 100

#Calcular a Percentagem Total de Valores Atípicos Pré-Truncamento aos Limites Inferior e Superior:
percentagem_total_valores_atípicos = (contador_número_total_valores_atípicos / len(conjunto_dados)) * 100

#Calcular a Percentagem Total de Valores Atípicos Pós-Truncamento aos Limites Inferior e Superior:
percentagem_total_valores_atípicos_pós_truncamento = (contador_número_total_valores_atípicos_pós_truncamento / (len(conjunto_dados) * len(conjunto_dados.columns))) * 100

#Mostrar a Percentagem de Valores Atípicos para cada Variável - Intervalo Interquartil:
print("Percentagem de Valores Atípicos para cada Variável - Intervalo Interquartil:")
print(f"{chr(10)}Variável:{' ' * 21}Percentagem de Valores Atípicos - Intervalo Interquartil:\n")
for variável, percentagem in percentagem_valores_atípicos_variável.items():
    print(f"{variável:<30}{percentagem:.2f}%")

#Mostrar a Percentagem Total de Valores Atípicos do Conjunto de Dados - Intervalo Interquartil:
print(f"{chr(10)}Percentagem Total de Valores Atípicos do Conjunto de Dados - Intervalo Interquartil: {percentagem_total_valores_atípicos:.2f}%")

#Mostrar a Percentagem de Valores Atípicos para cada Variável Pós-Truncamento aos Limites Inferior e Superior - Intervalo Interquartil::
print("\n\nPercentagem de Valores Atípicos para cada Variável Pós-Truncamento aos Limites Inferior e Superior - Intervalo Interquartil:")
print(f"{chr(10)}Variável:{' ' * 21}Percentagem de Valores Atípicos Pós-Truncamento aos Limites Inferior e Superior - Intervalo Interquartil:\n")
for variável, percentagem in percentagem_valores_atípicos_variável_pós_truncamento.items():
    print(f"{variável:<30}{percentagem:.2f}%")

#Mostrar a Percentagem Total de Valores Atípicos do Conjunto de Dados Pós-Truncamento aos Limites Inferior e Superior:
print(f"{chr(10)}Percentagem Total de Valores Atípicos do Conjunto de Dados Pós-Truncamento aos Limites Inferior e Superior - Intervalo Interquartil: {percentagem_total_valores_atípicos_pós_truncamento:.2f}%")

#Guardar o Conjunto de Dados Sem Valores Atípicos Pós-Truncamento aos Limites Inferior e Superior:
conjunto_dados.to_csv('conjunto_dados_sem_valores_atípicos.csv', index=False)
print("\n\nO Conjunto de Dados Sem Valores Atípicos Pós-Truncamento aos Limites Inferior e Superior Foi Guardado com Sucesso como conjunto_dados_sem_valores_atípicos.csv")
```

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar a Biblioteca Numerical Python:
import numpy as np

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_sem_valores_atípicos.csv')

#Calcular a Matriz de Correlação do Conjunto de Dados:
matriz_correlação_conjunto_dados = conjunto_dados.corr()

#Substituir os Valores da Diagonal Principal da Matriz de Correlação do Conjunto de Dados por 0:
np.fill_diagonal(matriz_correlação_conjunto_dados.values, 0)

#Encontrar os Pares de Variáveis com uma Correlação Muito Forte (Superior a 0.90):
pares_variáveis_correlação_muito_forte = matriz_correlação_conjunto_dados.stack().loc[lambda x: np.abs(x) > 0.90]

#Criar um Conjunto para Armazenar os Pares de Variáveis com uma Correlação Muito Forte:
conjunto_pares_variáveis_correlação_muito_forte = set()

#Iterar sobre os Pares de Variáveis com uma Correlação Muito Forte:
for (variável_redundante_1, variável_redundante_2), variável in pares_variáveis_correlação_muito_forte.items():

    #Adicionar os Pares de Variáveis com uma Correlação Muito Forte ao Conjunto:
    conjunto_pares_variáveis_correlação_muito_forte.add(tuple(sorted([variável_redundante_1, variável_redundante_2])))

#Mostrar os Pares de Variáveis com uma Correlação Muito Forte (Superior a 0.90):
print("Pares de Variáveis com uma Correlação Muito Forte (Superior a 0.90):\n")

#Iterar sobre os Pares de Variáveis com uma Correlação Muito Forte:
for variável_redundante_1, variável_redundante_2 in sorted(conjunto_pares_variáveis_correlação_muito_forte):

    #Mostrar os Pares de Variáveis com uma Correlação Muito Forte e Suas Correlações:
    print(f"Par de Variáveis {variável_redundante_1} e {variável_redundante_2} - Correlação = {matriz_correlação_conjunto_dados.loc[variável_redundante_1, variável_redundante_2]:.2f}")

#Identificar as Variáveis Redundantes dos Pares de Variáveis com uma Correlação Muito Forte:
variáveis_redundantes = set(variável_redundante for pares_variáveis_correlação_muito_forte in conjunto_pares_variáveis_correlação_muito_forte for variável_redundante in pares_variáveis_correlação_muito_forte)
```

<H5>3.2. Algoritmos de Clustering:</H5>

<H5>3.2.1. Algoritmo de Clustering K-Means:</H5>

<H5>3.2.1.1. Escolha do Número Ótimo de Clusters (k) para Maximizar o Índice de Silhueta Global:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar o Submódulo Cluster da Biblioteca Scikit-Learn:
from sklearn import cluster

#Importar o Submódulo Metrics da Biblioteca Scikit-Learn:
from sklearn import metrics

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv("conjunto_dados_clustering.csv")

#Obter as Variáveis de Entrada do Conjunto de Dados:
variáveis_entrada = conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1)

#Iniciar o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 0

#Iniciar o Maior Índice de Silhueta:
maior_índice_silhueta = -1

#Iterar sobre o Intervalo de Valores do Número de Clusters (k):
for k in range(2, 16):
    
    #Aplicar o Algoritmo de Clustering K-Means com k Clusters para Ajustar o Modelo:
    k_means = cluster.KMeans(n_clusters=k, random_state=1, init='k-means++')
    
    #Obter os Rótulos dos Clusters:
    rótulos_clusters = k_means.fit_predict(variáveis_entrada)

    #Calcular o Índice de Silhueta Global:
    índice_silhueta_global = metrics.silhouette_score(variáveis_entrada, rótulos_clusters)

    #Mostrar o Índice de Silhueta Global para k Clusters:
    print(f"\nÍndice de Silhueta Global para {k} Clusters: {índice_silhueta_global:.3f}")
    
    #Se o Índice de Silhueta Global for Superior ao Maior Índice de Silhueta Atual:
    if índice_silhueta_global > maior_índice_silhueta:

        #Atualizar o Maior Índice de Silhueta:
        maior_índice_silhueta = índice_silhueta_global

        #Atualizar o Número Ótimo de Clusters (k):
        número_ótimo_clusters_k = k

#Mostrar o Número Ótimo de Clusters (k) para o Maior Índice de Silhueta Global:
print(f"\n\nNúmero Ótimo de Clusters (k) = {número_ótimo_clusters_k}, Índice de Silhueta Global de {maior_índice_silhueta:.3f}")
```

<H5>3.2.1.2. Centróides dos Clusters:</H5>

<H5>3.2.1.2.1. Treino do Modelo de K-Means e Visualização dos Centróides dos Clusters:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar o Módulo de Clustering da Biblioteca Scikit-Learn:
from sklearn import cluster

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv("conjunto_dados_clustering.csv")

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1, errors='ignore')

#Definir o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 14

#Definir o Algoritmo de Clustering K-Means:
algoritmo_k_means = cluster.KMeans(n_clusters=número_ótimo_clusters_k, random_state=1)

#Treinar o Modelo de K-Means:
modelo_k_means = algoritmo_k_means.fit(variáveis_entrada)

#Obter os Centróides dos Clusters:
centróides_clusters = modelo_k_means.cluster_centers_

#Criar um Conjunto de Dados com os Centróides dos Clusters:
conjunto_dados_k_means_centróides_clusters = pd.DataFrame(data=centróides_clusters, columns=variáveis_entrada.columns).round(2)

#Configurações da Exibição de Resultados - Largura Máxima de Exibição do DataFrame:
pd.set_option('display.width', 1000)

#Configurações da Exibição de Resultados - Largura Máxima de Exibição de Cada Célula do DataFrame:
pd.set_option('display.max_colwidth', None)

#Configurações da Exibição de Resultados - Mostrar Todas as Variáveis do Conjunto de Dados numa única Linha:
pd.set_option('display.max_columns', None)

#Mostrar o Conjunto de Dados com os Centróides dos Clusters:
print("Conjunto de Dados com os Centróides dos Clusters:\n")
print(conjunto_dados_k_means_centróides_clusters)

#Guardar o Conjunto de Dados com os Centróides dos Clusters:
conjunto_dados_k_means_centróides_clusters.to_csv("conjunto_dados_clustering_k_means_centróides_clusters.csv", index=False)
print("\n\nO Conjunto de Dados com os Centróides dos Clusters Foi Guardado com Sucesso em conjunto_dados_clustering_k_means_centróides_clusters.csv")
```

<H5>3.2.1.2.2. Cálculo dos Centróides dos Clusters nas Componentes Principais 1 e 2 utilizando a Análise de Componentes Principais e o Modelo de K-Means:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar o Módulo de Clustering da Biblioteca Scikit-Learn:
from sklearn import cluster

#A partir do Submódulo Preprocessing da Biblioteca Scikit-Learn, Importar a Classe StandardScaler:
from sklearn.preprocessing import StandardScaler

#A partir do Submódulo Decomposition da Biblioteca Scikit-Learn, Importar a Classe PCA:
from sklearn.decomposition import PCA

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_clustering.csv')

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(columns=['Receita_Total_Discretizada'])

#Definir o Normalizador de Escala Z-Score:
normalizador_escala_z_score = StandardScaler()

#Normalizar a Escala das Variáveis de Entrada com o Normalizador Z-Score:
variáveis_entrada_normalizadas_z_score = normalizador_escala_z_score.fit_transform(variáveis_entrada)

#Definir a Análise de Componentes Principais para 2 Componentes Principais:
análise_componentes_principais = PCA(n_components=2)

#Aplicar a Análise de Componentes Principais às Variáveis de Entrada Normalizadas:
dados_transformados_componentes_principais = análise_componentes_principais.fit_transform(variáveis_entrada_normalizadas_z_score)

#Definir o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 14

#Definir o Algoritmo de Clustering K-Means:
algoritmo_k_means = cluster.KMeans(n_clusters=número_ótimo_clusters_k, random_state=1)

#Treinar o Modelo de K-Means nos Dados Transformados pelas Componentes Principais 1 e 2:
modelo_k_means = algoritmo_k_means.fit(dados_transformados_componentes_principais)

#Obter os Centróides dos Clusters nas Componentes Principais 1 e 2:
centróides_clusters_componentes_principais = modelo_k_means.cluster_centers_

#Criar um Conjunto de Dados com os Centróides dos Clusters nas Componentes Principais 1 e 2:
conjunto_dados_clustering_k_means_centróides_clusters_componentes_principais = pd.DataFrame(data=centróides_clusters_componentes_principais, columns=['Componente Principal 1:', 'Componente Principal 2:']).round(2)

#Configurações da Exibição de Resultados - Largura Máxima de Exibição do DataFrame:
pd.set_option('display.width', 1000)

#Configurações da Exibição de Resultados - Largura Máxima de Exibição de Cada Célula do DataFrame:
pd.set_option('display.max_colwidth', None)

#Configurações da Exibição de Resultados - Mostrar Todas as Variáveis do Conjunto de Dados numa única Linha:
pd.set_option('display.max_columns', None)

#Mostrar o Conjunto de Dados com os Centróides dos Clusters nas Componentes Principais 1 e 2:
print("Centróides dos Clusters nas Componentes Principais 1 e 2:\n")
print(conjunto_dados_clustering_k_means_centróides_clusters_componentes_principais)

#Guardar o Conjunto de Dados com os Centróides dos Clusters nas Componentes Principais 1 e 2:
conjunto_dados_clustering_k_means_centróides_clusters_componentes_principais.to_csv("conjunto_dados_clustering_k_means_centróides_clusters_componentes_principais.csv", index=False)
print("\n\nOs Centróides dos Clusters nas Componentes Principais 1 e 2 Foram Guardados com Sucesso em conjunto_dados_clustering_k_means_centróides_clusters_componentes_principais.csv")
```

<H5>3.2.1.5. Avaliação do Modelo de K-Means atráves dos Índices Externos - Índice de Pureza e Índice de Rand:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar a Biblioteca Numerical Python:
import numpy as np

#Importar o Módulo de Clustering da Biblioteca Scikit-Learn:
from sklearn import cluster

#Importar as Métricas de Avaliação da Biblioteca Scikit-Learn:
from sklearn import metrics

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv("conjunto_dados_clustering.csv")

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1)

#Definir a Variável Alvo:
variável_alvo = conjunto_dados['Receita_Total_Discretizada']

#Definir o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 14

#Definir o Algoritmo de Clustering K-Means:
algoritmo_k_means = cluster.KMeans(n_clusters=número_ótimo_clusters_k, random_state=1)

#Treinar o Modelo de K-Means com as Variáveis de Entrada:
modelo_k_means = algoritmo_k_means.fit(variáveis_entrada)

#Obter os Rótulos dos Clusters:
rótulos_clusters = modelo_k_means.labels_

#Definir a Função do Índice de Pureza:
def função_índice_pureza(Y_verdadeiro, Y_previsto):
    
    #Calcular a Matriz de Contingência:
    matriz_contingência = metrics.cluster.contingency_matrix(Y_verdadeiro, Y_previsto)
    
    #Devolver o Índice de Pureza Calculado a partir da Matriz de Contingência:
    return np.sum(np.amax(matriz_contingência, axis=0)) / np.sum(matriz_contingência)

#Calcular o Índice de Pureza:
índice_pureza = função_índice_pureza(variável_alvo, rótulos_clusters)

#Mostrar o Índice de Pureza:
print(f"Índice de Pureza para {número_ótimo_clusters_k} Clusters: {índice_pureza:.2f}")

#Calcular o Índice de Rand:
índice_rand_ajustado = metrics.adjusted_rand_score(variável_alvo, rótulos_clusters)

#Mostrar o Índice de Rand:
print(f"\nÍndice de Rand para {número_ótimo_clusters_k} Clusters: {índice_rand_ajustado:.2f}")
```

<H5>3.2.1.8. Cálculo dos Centróides, Medianas, Medóides, Diferenças e Dispersão Média para cada Cluster:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#A partir do Submódulo Preprocessing da Biblioteca Scikit-Learn, importar a Classe StandardScaler:
from sklearn.preprocessing import StandardScaler

#A partir do Submódulo Decomposition da Biblioteca Scikit-Learn, importar a Classe PCA:
from sklearn.decomposition import PCA

#A partir do Submódulo Cluster da Biblioteca Scikit-Learn, importar a Classe KMeans:
from sklearn.cluster import KMeans

#Importar a Biblioteca Numerical Python:
import numpy as np

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_clustering.csv')

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(columns=['Receita_Total_Discretizada'])

#Definir o Normalizador de Escala Z-Score:
normalizador_escala_z_score = StandardScaler()

#Normalizar as Variáveis de Entrada com Z-Score:
variáveis_entrada_normalizadas_z_score = normalizador_escala_z_score.fit_transform(variáveis_entrada)

#Definir a Análise de Componentes Principais para as 2 Componentes Principais de Topo:
análise_componentes_principais = PCA(n_components=2)

#Aplicar a Análise de Componentes Principais nas Variáveis Normalizadas:
componentes_principais = análise_componentes_principais.fit_transform(variáveis_entrada_normalizadas_z_score)

#Definir o Número Ótimo de Clusters:
número_ótimo_clusters = 14

#Definir o Algoritmo de Clustering K-Means:
algoritmo_hierárquico = KMeans(n_clusters=número_ótimo_clusters, random_state=0)

#Aplicar o Algoritmo de Clustering para Prever os Clusters:
clusters = algoritmo_hierárquico.fit_predict(componentes_principais)

#Criar um Conjunto de Dados para Armazenar os Resultados:
conjunto_dados_clustering_hierárquico_resumo = pd.DataFrame(columns=[
    'Centróide', 'Mediana', 'Medoide',
    'Diferença_Centroide_Mediana', 'Diferença_Centroide_Medoide',
    'Diferença_Mediana_Medoide', 'Dispersão_Média'])

#Iterar sobre cada Cluster:
for cluster in range(número_ótimo_clusters):

    #Obter os Dados de cada Cluster:
    cluster_componentes = componentes_principais[clusters == cluster]

    #Calcular o Centróide:
    centroide = np.mean(cluster_componentes)

    #Calcular a Mediana:
    mediana = np.median(cluster_componentes)

    #Calcular o Medoide:
    distâncias = np.linalg.norm(cluster_componentes - mediana, axis=1)
    índice_medoide = np.argmin(distâncias)
    medoide = np.mean(cluster_componentes[índice_medoide])

    #Calcular as Diferenças entre o Centróide, Mediana e Medoide:
    diferença_centroide_mediana = centroide - mediana
    diferença_centroide_medoide = centroide - medoide
    diferença_mediana_medoide = mediana - medoide

    #Calcular a Dispersão Média:
    dispersão_média = np.mean(np.abs(cluster_componentes - mediana))

    #Adicionar ao Conjunto de Dados de Resumo:
    conjunto_dados_clustering_hierárquico_resumo.loc[cluster] = [
        centroide, mediana, medoide,
        diferença_centroide_mediana, diferença_centroide_medoide,
        diferença_mediana_medoide, dispersão_média]

#Configurações da Exibição de Resultados - Largura Máxima de Exibição do DataFrame:
pd.set_option('display.width', 1000)

#Configurações da Exibição de Resultados - Largura Máxima de Exibição de Cada Célula do DataFrame:
pd.set_option('display.max_colwidth', None)

#Configurações da Exibição de Resultados - Mostrar Todas as Variáveis do Conjunto de Dados numa única Linha:
pd.set_option('display.max_columns', None)

#Arredondar os Resultados para 2 Casas Decimais:
conjunto_dados_clustering_hierárquico_resumo = conjunto_dados_clustering_hierárquico_resumo.round(2)

#Mostrar os Resultados dos Centróides, Medianas, Medoides, Diferenças e Dispersões Médias para cada Cluster - Algoritmo de Clustering K-Means:
print("Centróide, Mediana, Medoide, Diferenças e Dispersão Média para cada Cluster - Algoritmo de Clustering K-Means:\n")
print(conjunto_dados_clustering_hierárquico_resumo)

#Guardar os Resultados dos Centróides, Medianas, Medoides, Diferenças e Dispersões Médias para cada Cluster - Algoritmo de Clustering K-Means:
conjunto_dados_clustering_hierárquico_resumo.to_csv(
    'conjunto_dados_clustering_k_means_medianas_medoides_dispersões_médias.csv', index=False, float_format='%.2f')
print("\n\nOs Resultados dos Centróides, Medianas, Medoides, Diferenças e Dispersões Médias para cada Cluster Foram Guardados com Sucesso em conjunto_dados_clustering_k_means_medianas_medoides_dispersões_médias.csv")
```

<H5>3.2.1.9. Importância das Variáveis para a Separação dos Clusters com base na Estatística F:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar a Biblioteca Numerical Python:
import numpy as np

#A partir do Submódulo Preprocessing da Biblioteca Scikit-Learn, Importar o StandardScaler:
from sklearn.preprocessing import StandardScaler

#Importar o Submódulo Pyplot da Biblioteca Matplotlib:
import matplotlib.pyplot as matplotlib

#A partir do Submódulo Feature_Selection da Biblioteca Scikit-Learn, Importar a Função f_classif:
from sklearn.feature_selection import f_classif

#Importar a Biblioteca Math:
import math

#A partir do Submódulo Display da Biblioteca IPython, Importar a Função Display:
from IPython.display import display

#Configurar a Exibição dos Números de Vírgula Flutuante com 2 Casas Decimais:
pd.options.display.float_format = '{:.2f}'.format

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_clustering_k_means_rótulos_clusters.csv')

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(columns=['Cluster'])

#Obter os Rótulos dos Clusters:
rótulos_clusters = conjunto_dados['Cluster']

#Iniciar o Normalizador de Escala Z-Score:
normalizador = StandardScaler()

#Normalizar as Variáveis de Entrada com o Normalizador Z-Score:
variáveis_normalizadas = normalizador.fit_transform(variáveis_entrada)

#Listar os Clusters Únicos Ordenados:
clusters_únicos = sorted(rótulos_clusters.unique())

#Criar Dicionários para Armazenar os Valores F e os DataFrames de Resultados por Cluster:
valores_f_dict = {}
df_valores_f_dict = {}

#Calcular a Estatística F para Cada Cluster:
for cluster in clusters_únicos:
    
    #Criar um Vetor Binário para o Cluster Atual:
    rótulos_binários = (rótulos_clusters == cluster).astype(int)
    
    #Calcular a Estatística F para Cada Variável:
    valor_f, p_valores = f_classif(variáveis_normalizadas, rótulos_binários)
    
    #Arredondar os Valores F para 2 Casas Decimais:
    valor_f = np.round(valor_f, 2)
    
    #Armazenar os Valores F no Dicionário:
    valores_f_dict[cluster] = valor_f
    
    #Criar um DataFrame com os Nomes das Variáveis e os Respetivos Valores F:
    df_f = pd.DataFrame({'Variável': variáveis_entrada.columns, 'Valor F': valor_f})
    
    #Ordenar as Variáveis em Ordem Decrescente de Importância:
    df_f = df_f.sort_values(by='Valor F', ascending=False).reset_index(drop=True)
    df_valores_f_dict[cluster] = df_f

#Calcular a Importância Global das Variáveis para os Clusters Multiclasse:
valor_f_global, p_valores_global = f_classif(variáveis_normalizadas, rótulos_clusters)
valor_f_global = np.round(valor_f_global, 2)
df_f_global = pd.DataFrame({
    'Variável': variáveis_entrada.columns,
    'Valor F': valor_f_global})
df_f_global = df_f_global.sort_values(by='Valor F', ascending=False).reset_index(drop=True)

#Mostrar as Tabelas com os Valores F para Cada Cluster:
for cluster in clusters_únicos:
    print(f"\nResultados para o Cluster {cluster}:\n")
    print(df_valores_f_dict[cluster])

#Mostrar a Tabela com os Valores F Globais:
print("\nResultados Globais Considerando Todos os Clusters:\n")
print(df_f_global)

#Definir o Número Total de Gráficos Um para Cada Cluster + 1 Gráfico Global:
total_graficos = len(clusters_únicos) + 1

#Configurar o Layout dos Subplots:
n_colunas = 4
n_linhas = math.ceil(total_graficos / n_colunas)

figura, eixos = matplotlib.subplots(n_linhas, n_colunas, figsize=(n_colunas * 5, n_linhas * 6), constrained_layout=True)
eixos = eixos.flatten()

#Ajustar o Espaçamento entre os Gráficos:
figura.subplots_adjust(hspace=0.7, wspace=0.7)

#Gerar os Gráficos de Barras Individuais para Cada Cluster:
for i, cluster in enumerate(clusters_únicos):
    eixo = eixos[i]
    df_f = df_valores_f_dict[cluster]
    
    #Criar o Gráfico de Barras para o Cluster Atual:
    barras = eixo.bar(df_f['Variável'], df_f['Valor F'], color='#2ca02c', edgecolor='#2ca02c')
    
    #Adicionar o Valor F Acima de Cada Barra:
    for barra in barras:
        altura = barra.get_height()
        eixo.text(barra.get_x() + barra.get_width() / 2, altura, f'{altura:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
                  ha='center', va='bottom', fontsize=8)
    
    eixo.set_title(f'Cluster {cluster}', fontsize=16)
    eixo.set_xlabel('Variável', fontsize=14)
    eixo.set_ylabel('Valor F', fontsize=14)
    eixo.tick_params(axis='x', rotation=45, labelsize=8)
    eixo.tick_params(axis='y', labelsize=8)
    
    #Ajustar o Limite Superior do Eixo Y:
    max_valor_f = np.max(df_f['Valor F'])
    
    #Calcular o Intervalo de Valores com base no Valor Máximo:
    intervalo = max(50, max_valor_f * 0.2)
    eixo.set_ylim(0, max_valor_f + intervalo)

#Gerar o Gráfico de Barras Global:
eixo_global = eixos[len(clusters_únicos)]
barras = eixo_global.bar(df_f_global['Variável'], df_f_global['Valor F'], color='#2ca02c', edgecolor='#2ca02c')
for barra in barras:
    altura = barra.get_height()
    eixo_global.text(barra.get_x() + barra.get_width() / 2, altura, f'{altura:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
                     ha='center', va='bottom', fontsize=8)
eixo_global.set_title('Global', fontsize=16)
eixo_global.set_xlabel('Variável\n', fontsize=14)
eixo_global.set_ylabel('Valor F', fontsize=14)
eixo_global.tick_params(axis='x', rotation=45, labelsize=8)
eixo_global.tick_params(axis='y', labelsize=8)

#Ajustar o Limite Superior do Eixo Y para o Gráfico Global:
max_valor_f_global = np.max(df_f_global['Valor F'])
intervalo_global = max(50, max_valor_f_global * 0.2)
eixo_global.set_ylim(0, max_valor_f_global + intervalo_global)

#Ocultar os Subgráficos Vazios:
for j in range(total_graficos, len(eixos)):
    eixos[j].set_visible(False)

#Definir o Título Geral da Figura:
figura.suptitle('Conjunto de Dados 1 - Algoritmo de Clustering K-Means\n\nImportância das Variáveis para a Separação dos 14 Clusters com base no Valor F\n', fontsize=20)

#Mostrar o Gráfico:
matplotlib.show()
```

<H5>3.2.2.1. Escolha do Número Ótimo de Clusters (k) e do Melhor Critério de Ligação para Maximizar o Índice de Silhueta Global:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar o Submódulo Cluster da Biblioteca Scikit-Learn:
from sklearn import cluster

#Importar o Submódulo Metrics da Biblioteca Scikit-Learn:
from sklearn import metrics

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv("conjunto_dados_clustering.csv")

#Obter as Variáveis de Entrada do Conjunto de Dados:
variáveis_entrada = conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1)

#Iniciar o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 0

#Definir os Tipos de Ligação:
tipos_ligação = ['Simples', 'Média', 'Completa', 'Ward']

#Iniciar o Maior Índice de Silhueta:
maior_índice_silhueta = -1

#Iterar sobre o Intervalo de Valores do Número de Clusters (k):
for k in range(10, 16):
    
    #Iterar sobre os Tipos de Ligação:
    for tipo_ligação in tipos_ligação:
        
        #Traduzir os Nomes dos Tipos de Ligação:
        if tipo_ligação == 'Simples':
            tipo_ligação_inglês = 'single'
        
        elif tipo_ligação == 'Média':
            tipo_ligação_inglês = 'average'
        
        elif tipo_ligação == 'Completa':
            tipo_ligação_inglês = 'complete'
        
        else:
            tipo_ligação_inglês = 'Ward'
        
        #Aplicar o Algoritmo de Clustering Aglomerativo Hierárquico com k Clusters e Tipo de Ligação:
        aglomerativo_hierárquico = cluster.AgglomerativeClustering(n_clusters=k, linkage=tipo_ligação_inglês)
        
        #Obter os Rótulos dos Clusters:
        rótulos_clusters = aglomerativo_hierárquico.fit_predict(variáveis_entrada)

        #Calcular o Índice de Silhueta Global:
        índice_silhueta_global = metrics.silhouette_score(variáveis_entrada, rótulos_clusters)

        #Mostrar o Índice de Silhueta Global para k Clusters e Tipo de Ligação:
        print(f"\nÍndice de Silhueta Global para {k} Clusters e Ligação '{tipo_ligação}': {índice_silhueta_global:.4f}")
        
        #Se o Índice de Silhueta Global for Superior ao Maior Índice de Silhueta Atual:
        if índice_silhueta_global > maior_índice_silhueta:

            #Atualizar o Maior Índice de Silhueta:
            maior_índice_silhueta = índice_silhueta_global

            #Atualizar o Número Ótimo de Clusters (k):
            número_ótimo_clusters_k = k

            #Atualizar o Tipo de Ligação:
            melhor_tipo_ligação = tipo_ligação

#Mostrar o Número Ótimo de Clusters (k) e Tipo de Ligação para o Maior Índice de Silhueta Global:
print(f"\n\nNúmero Ótimo de Clusters (k) = {número_ótimo_clusters_k}, Tipo de Ligação = '{melhor_tipo_ligação}', Índice de Silhueta Global de {maior_índice_silhueta:.4f}")
```

<H5>3.2.2.3. Avaliação do Modelo de Clustering Hierárquico Aglomerativo atráves do Índice Interno de Davis-Bouldin:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar o Módulo de Clustering da Biblioteca Scikit-Learn:
from sklearn import cluster

#A partir do Submódulo Metrics da Biblioteca Scikit-Learn, Importar a Função para Calcular as Distâncias entre os Registos:
from sklearn.metrics import pairwise_distances

#Importar a Biblioteca Numerical Python:
import numpy as np

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv("conjunto_dados_clustering.csv")

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1, errors='ignore')

#Definir o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 12

#Definir o Tipo de Ligação:
tipo_ligação = 'ward'

#Definir o Algoritmo de Clustering Hierárquico Aglomerativo:
algoritmo_aglomerativo = cluster.AgglomerativeClustering(n_clusters=número_ótimo_clusters_k, linkage=tipo_ligação)

#Obter os Rótulos dos Clusters:
rótulos_clusters = algoritmo_aglomerativo.fit_predict(variáveis_entrada)

#Definir a Função para Calcular o Índice de Davies-Bouldin:
def função_calcular_índice_davis_bouldin(X, rótulos):
    
    #Contar o Número de Clusters:
    número_clusters = len(np.unique(rótulos))
    
    #Iniciar o Índice de Davies-Bouldin:
    índice_davis_bouldin = 0
    
    #Calcular as distâncias entre os pontos:
    distâncias = pairwise_distances(X)
    
    #Para cada Par de clusters:
    for i in range(número_clusters):
        for j in range(i+1, número_clusters):
            
            #Selecionar os Pontos de cada Cluster:
            pontos_cluster_i = X[rótulos == i]
            pontos_cluster_j = X[rótulos == j]
            
            #Calcular o Ponto Médio de cada Cluster:
            ponto_médio_i = np.mean(pontos_cluster_i, axis=0)
            ponto_médio_j = np.mean(pontos_cluster_j, axis=0)
            
            #Calcular a Distância entre os Pontos Médios:
            distância_pontos_médios = np.linalg.norm(ponto_médio_i - ponto_médio_j)
            
            #Calcular a Dispersão de cada Cluster:
            dispersão_i = np.mean(pairwise_distances(pontos_cluster_i, [ponto_médio_i]))
            dispersão_j = np.mean(pairwise_distances(pontos_cluster_j, [ponto_médio_j]))
            
            #Calcular a Medida de Separação e Compactação:
            índice_davis_bouldin += (dispersão_i + dispersão_j) / distância_pontos_médios
    
    #Calcular o Índice de Davies-Bouldin:
    índice_davis_bouldin = índice_davis_bouldin / (número_clusters * (número_clusters - 1) / 2)
    
    #Devolver o Valor do Índice de Davies-Bouldin:
    return índice_davis_bouldin

#Calcular o Índice de Davies-Bouldin:
índice_davies_bouldin = função_calcular_índice_davis_bouldin(variáveis_entrada.values, rótulos_clusters)

#Mostrar o Índice de Davies-Bouldin:
print(f"Índice de Davies-Bouldin (DBI): {índice_davies_bouldin:.3f}")
```

<H5>3.2.2.5. Avaliação do Modelo de Clustering Hierárquico Aglomerativo atráves dos Índices Externos - Índice de Pureza e Índice de Rand:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar a Biblioteca Numerical Python:
import numpy as np

#Importar o Módulo de Clustering da Biblioteca Scikit-Learn:
from sklearn import cluster

#Importar as Métricas de Avaliação da Biblioteca Scikit-Learn:
from sklearn import metrics

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv("conjunto_dados_clustering.csv")

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1)

#Definir a Variável Alvo:
variável_alvo = conjunto_dados['Receita_Total_Discretizada']

#Definir o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 12

#Definir o Tipo de Ligação:
tipo_ligação='ward'

#Definir o Algoritmo de Clustering Hierárquico Aglomerativo com Ligação Ward:
algoritmo_hierárquico = cluster.AgglomerativeClustering(n_clusters=número_ótimo_clusters_k, linkage=tipo_ligação)

#Treinar o Modelo de Clustering Hierárquico com as Variáveis de Entrada:
modelo_hierárquico = algoritmo_hierárquico.fit(variáveis_entrada)

#Obter os Rótulos dos Clusters:
rótulos_clusters = modelo_hierárquico.labels_

#Definir a Função do Índice de Pureza:
def função_índice_pureza(Y_verdadeiro, Y_previsto):
    
    #Calcular a Matriz de Contingência:
    matriz_contingência = metrics.cluster.contingency_matrix(Y_verdadeiro, Y_previsto)
    
    #Devolver o Índice de Pureza Calculado a partir da Matriz de Contingência:
    return np.sum(np.amax(matriz_contingência, axis=0)) / np.sum(matriz_contingência)

#Calcular o Índice de Pureza:
índice_pureza = função_índice_pureza(variável_alvo, rótulos_clusters)

#Mostrar o Índice de Pureza:
print(f"Índice de Pureza para {número_ótimo_clusters_k} Clusters: {índice_pureza:.2f}")

#Calcular o Índice de Rand:
índice_rand_ajustado = metrics.adjusted_rand_score(variável_alvo, rótulos_clusters)

#Mostrar o Índice de Rand:
print(f"\nÍndice de Rand para {número_ótimo_clusters_k} Clusters: {índice_rand_ajustado:.2f}")
```

<H5>3.2.2.6. Obtenção dos Rótulos dos Clusters:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar o Submódulo Dendrograma da Biblioteca SciPy:
from scipy.cluster.hierarchy import linkage, fcluster

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv("conjunto_dados_clustering.csv")

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1, errors='ignore')

#Definir o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 12

#Definir o Tipo de Ligação:
tipo_ligação = 'ward'

#Definir a Medida de Distância:
medida_distância = 'euclidean'

#Criar a Matriz de Distâncias com o Método de Ligação Ward e a Distância Euclidiana:
matriz_distâncias = linkage(variáveis_entrada, method=tipo_ligação, metric=medida_distância)

#Obter os Rótulos dos Clusters:
rótulos_clusters = fcluster(matriz_distâncias, t=número_ótimo_clusters_k, criterion='maxclust')

#Adicionar os Rótulos dos Clusters ao Conjunto de Dados:
conjunto_dados['Cluster'] = rótulos_clusters

#Mostrar os 5 Primeiros Registos do Conjunto de Dados com os Rótulos dos Clusters:
print("Conjunto de Dados com os Rótulos dos Clusters - 5 Primeiros Registos:\n")
print(conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1).round(2).head().to_string(index=False))

#Guardar o Conjunto de Dados com os Rótulos dos Clusters:
conjunto_dados.drop(['Receita_Total_Discretizada'], axis=1, errors='ignore').round(2).to_csv("conjunto_dados_clustering_hierárquico_aglomerativo_rótulos_clusters.csv", index=False)
print("\n\nO Conjunto de Dados com os Rótulos dos Clusters Foi Guardado com Sucesso em conjunto_dados_clustering_hierárquico_aglomerativo_rótulos_clusters.csv")
```

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar a Biblioteca Numerical Python:
import numpy as np

#Importar o Submódulo Pyplot da Biblioteca Matplotlib:
import matplotlib.pyplot as plt

#A partir do Submódulo Colors da Biblioteca Matplotlib, Importar a Classe LinearSegmentedColormap:
from matplotlib.colors import LinearSegmentedColormap

#A partir do Submódulo Preprocessing da Biblioteca Scikit-Learn, Importar a Classe StandardScaler:
from sklearn.preprocessing import StandardScaler

#A partir do Submódulo Decomposition da Biblioteca Scikit-Learn, Importar a Classe PCA:
from sklearn.decomposition import PCA

#A partir do Submódulo Cluster da Biblioteca Scikit-Learn, Importar a Classe AgglomerativeClustering:
from sklearn.cluster import AgglomerativeClustering

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_clustering.csv')

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(columns=['Receita_Total_Discretizada'])

#Definir o Normalizador de Escala Z-Score:
normalizador_escala_z_score = StandardScaler()

#Normalizar a Escala das Variáveis de Entrada com o Normalizador Z-Score:
variáveis_entrada_normalizadas_z_score = normalizador_escala_z_score.fit_transform(variáveis_entrada)

#Definir a Análise de Componentes Principais para 2 Componentes Principais:
análise_componentes_principais = PCA(n_components=2)

#Aplicar a Análise de Componentes Principais às Variáveis de Entrada Normalizadas:
componentes_principais = análise_componentes_principais.fit_transform(variáveis_entrada_normalizadas_z_score)

#Definir o Número Ótimo de Clusters (k):
número_ótimo_clusters_k = 12

#Definir o Tipo de Ligação:
tipo_ligação = 'ward'

#Definir o Algoritmo de Clustering Hierárquico Aglomerativo com Ligação Ward:
algoritmo_hierárquico = AgglomerativeClustering(n_clusters=número_ótimo_clusters_k, linkage=tipo_ligação)

#Aplicar o Algoritmo de Clustering Hierárquico Aglomerativo para Prever os Clusters com Base nas Componentes Principais:
clusters = algoritmo_hierárquico.fit_predict(componentes_principais)

#Configurações da Escala de Cores - Cor Amarelo até Cor Roxo para 12 Clusters:
escala_cores = LinearSegmentedColormap.from_list("amarelo_roxo", ["#ffff00", "#800080"], N=número_ótimo_clusters_k)

#Configurações do Gráfico 2D - Figura - Tamanho:
plt.figure(figsize=(10, 6))

#Distribuição dos Clusters nas Componentes Principais - Configurações dos Pontos - Escala de Cores, Marcador e Tamanho do Marcador:
gráfico_dispersão = plt.scatter(componentes_principais[:, 0], componentes_principais[:, 1], c=clusters, cmap=escala_cores, marker='o', s=50)

#Iterar sobre os Índices dos Clusters:
for índice_cluster in range(número_ótimo_clusters_k):

    #Calcular a Posição Central para cada Rótulo:
    posição_horizontal = np.mean(componentes_principais[clusters == índice_cluster, 0])
    posição_vertical = np.mean(componentes_principais[clusters == índice_cluster, 1])
   
    #Adicionar o Rótulo para cada Cluster:
    plt.text(posição_horizontal, posição_vertical, f'Cluster {índice_cluster}', color='black', fontsize=12, ha='center')

    #Calcular os Centrósides das Componentes Principais para cada Cluster:
    centroide_componente_1 = np.mean(componentes_principais[clusters == índice_cluster, 0])
    centroide_componente_2 = np.mean(componentes_principais[clusters == índice_cluster, 1])
   
    #Adicionar o Símbolo do Centróide:
    plt.scatter(centroide_componente_1, centroide_componente_2, color='red', marker='X', s=200, label=f'Centróides' if índice_cluster == 0 else "")

#Configurações do Gráfico 2D - Título - Designação e Tamanho:
plt.title('Conjunto de Dados 1 - Algoritmo de Clustering Hierárquico Aglomerativo\n\nRepresentação dos 12 Clusters e respetivos Centróides no Espaço Bidimensional\ndas Componentes Principais 1 e 2 utilizando a Ligação de Ward e a Distância Euclidiana\n', fontsize=16)

#Configurações do Eixo X - Título - Designação e Tamanho:
plt.xlabel('Componente Principal 1', fontsize=14)

#Configurações do Eixo Y - Título - Designação e Tamanho:
plt.ylabel('Componente Principal 2', fontsize=14)

#Configurações da Barra de Cores - Designação, Tamanho, Espaçamento e Rotação:
plt.colorbar(gráfico_dispersão).set_label('Clusters', fontsize=12, labelpad=30, rotation=0)

#Mostrar o Gráfico 2D:
plt.legend()
plt.show()
```

<H5>3.2.2.8. Cálculo dos Centróides, Medianas, Medóides, Diferenças e Dispersão Média para cada Cluster:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#A partir do Submódulo Preprocessing da Biblioteca Scikit-Learn, importar a Classe StandardScaler:
from sklearn.preprocessing import StandardScaler

#A partir do Submódulo Decomposition da Biblioteca Scikit-Learn, importar a Classe PCA:
from sklearn.decomposition import PCA

#A partir do Submódulo Cluster da Biblioteca Scikit-Learn, importar a Classe AgglomerativeClustering:
from sklearn.cluster import AgglomerativeClustering

#Importar a Biblioteca Numerical Python:
import numpy as np

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_clustering.csv')

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(columns=['Receita_Total_Discretizada'])

#Definir o Normalizador de Escala Z-Score:
normalizador_escala_z_score = StandardScaler()

#Normalizar as Variáveis de Entrada com Z-Score:
variáveis_entrada_normalizadas_z_score = normalizador_escala_z_score.fit_transform(variáveis_entrada)

#Definir a Análise de Componentes Principais para as 2 Componentes Principais de Topo:
análise_componentes_principais = PCA(n_components=2)

#Aplicar a Análise de Componentes Principais nas Variáveis Normalizadas:
componentes_principais = análise_componentes_principais.fit_transform(variáveis_entrada_normalizadas_z_score)

#Definir o Número Ótimo de Clusters:
número_ótimo_clusters = 12

#Definir o Tipo de Ligação:
tipo_ligação = 'ward'

#Definir o Algoritmo de Clustering Hierárquico Aglomerativo com ligação Ward:
algoritmo_hierárquico = AgglomerativeClustering(n_clusters=número_ótimo_clusters, linkage=tipo_ligação)

#Aplicar o Algoritmo de Clustering para Prever os Clusters:
clusters = algoritmo_hierárquico.fit_predict(componentes_principais)

#Criar um Conjunto de Dados para Armazenar os Resultados:
conjunto_dados_clustering_hierárquico_resumo = pd.DataFrame(columns=[
    'Centróide', 'Mediana', 'Medoide',
    'Diferença_Centroide_Mediana', 'Diferença_Centroide_Medoide',
    'Diferença_Mediana_Medoide', 'Dispersão_Média'])

#Iterar sobre cada Cluster:
for cluster in range(número_ótimo_clusters):

    #Obter os Dados de cada Cluster:
    cluster_componentes = componentes_principais[clusters == cluster]

    #Calcular o Centróide:
    centroide = np.mean(cluster_componentes)

    #Calcular a Mediana:
    mediana = np.median(cluster_componentes)

    #Calcular o Medoide:
    distâncias = np.linalg.norm(cluster_componentes - mediana, axis=1)
    índice_medoide = np.argmin(distâncias)
    medoide = np.mean(cluster_componentes[índice_medoide])

    #Calcular as Diferenças entre o Centróide, Mediana e Medoide:
    diferença_centroide_mediana = centroide - mediana
    diferença_centroide_medoide = centroide - medoide
    diferença_mediana_medoide = mediana - medoide

    #Calcular a Dispersão Média:
    dispersão_média = np.mean(np.abs(cluster_componentes - mediana))

    #Adicionar ao Conjunto de Dados de Resumo:
    conjunto_dados_clustering_hierárquico_resumo.loc[cluster] = [
        centroide, mediana, medoide,
        diferença_centroide_mediana, diferença_centroide_medoide,
        diferença_mediana_medoide, dispersão_média]

#Configurações da Exibição de Resultados - Largura Máxima de Exibição do DataFrame:
pd.set_option('display.width', 1000)

#Configurações da Exibição de Resultados - Largura Máxima de Exibição de Cada Célula do DataFrame:
pd.set_option('display.max_colwidth', None)

#Configurações da Exibição de Resultados - Mostrar Todas as Variáveis do Conjunto de Dados numa única Linha:
pd.set_option('display.max_columns', None)

#Arredondar os Resultados para 2 Casas Decimais:
conjunto_dados_clustering_hierárquico_resumo = conjunto_dados_clustering_hierárquico_resumo.round(2)

#Mostrar os Resultados dos Centróides, Medianas, Medoides, Diferenças e Dispersões Médias para cada Cluster - Algoritmo de Clustering Hierárquico Aglomerativo:
print("Centróide, Mediana, Medoide, Diferenças e Dispersão Média para cada Cluster - Algoritmo de Clustering Hierárquico Aglomerativo:\n")
print(conjunto_dados_clustering_hierárquico_resumo)

#Guardar os Resultados dos Centróides, Medianas, Medoides, Diferenças e Dispersões Médias para cada Cluster - Algoritmo de Clustering Hierárquico Aglomerativo:
conjunto_dados_clustering_hierárquico_resumo.to_csv('conjunto_dados_clustering_hierárquici_aglomerativo_medianas_medoides_dispersoes_medias.csv', index=False, float_format='%.2f')
print("\n\nOs Resultados dos Centróides, Medianas, Medoides, Diferenças e Dispersões Médias para cada Cluster Foram Guardados com Sucesso em conjunto_dados_clustering_hierárquici_aglomerativo_medianas_medoides_dispersoes_medias.csv")
```

<H5>3.2.2.9. Importância das Variáveis para a Separação dos Clusters com base no Valor F:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar a Biblioteca Numerical Python:
import numpy as np

#A partir do Submódulo Preprocessing da Biblioteca Scikit-Learn, Importar o StandardScaler:
from sklearn.preprocessing import StandardScaler

#Importar o Submódulo Pyplot da Biblioteca Matplotlib:
import matplotlib.pyplot as matplotlib

#A partir do Submódulo Feature_Selection da Biblioteca Scikit-Learn, Importar a Função f_classif:
from sklearn.feature_selection import f_classif

#Importar a Biblioteca Math:
import math

#A partir do Submódulo Display da Biblioteca IPython, Importar a Função Display:
from IPython.display import display

#Configurar a Exibição de Floats com 2 Casas Decimais:
pd.options.display.float_format = '{:.2f}'.format

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_clustering_hierárquico_aglomerativo_rótulos_clusters.csv')

#Obter as Variáveis de Entrada:
variáveis_entrada = conjunto_dados.drop(columns=['Cluster'])

#Obter os Rótulos dos Clusters:
rótulos_clusters = conjunto_dados['Cluster'] - 1

#Iniciar o Normalizador de Escala Z-Score:
normalizador = StandardScaler()

#Normalizar as Variáveis de Entrada com o Normalizador Z-Score:
variáveis_normalizadas = normalizador.fit_transform(variáveis_entrada)

#Listar os Clusters Únicos Ordenados:
clusters_únicos = sorted(rótulos_clusters.unique())

#Criar Dicionários para Armazenar os Valores F e os DataFrames de Resultados por Cluster:
valores_f_dict = {}
df_valores_f_dict = {}

#Calcular a Estatística F para Cada Cluster:
for cluster in clusters_únicos:
    
    #Criar um Vetor Binário para o Cluster Atual:
    rótulos_binários = (rótulos_clusters == cluster).astype(int)
    
    #Calcular a Estatística F para Cada Variável:
    valor_f, p_valores = f_classif(variáveis_normalizadas, rótulos_binários)
    
    #Arredondar os Valores F para 2 Casas Decimais:
    valor_f = np.round(valor_f, 2)
    
    #Armazenar os Valores F no Dicionário:
    valores_f_dict[cluster] = valor_f
    
    #Criar um DataFrame com os Nomes das Variáveis e os Respetivos Valores F:
    df_f = pd.DataFrame({'Variável': variáveis_entrada.columns, 'Valor F': valor_f})
    
    #Ordenar as Variáveis em Ordem Decrescente de Importância:
    df_f = df_f.sort_values(by='Valor F', ascending=False).reset_index(drop=True)
    df_valores_f_dict[cluster] = df_f

#Calcular a Importância Global das Variáveis para os Clusters Multiclasse:
valor_f_global, p_valores_global = f_classif(variáveis_normalizadas, rótulos_clusters)
valor_f_global = np.round(valor_f_global, 2)
df_f_global = pd.DataFrame({
    'Variável': variáveis_entrada.columns,
    'Valor F': valor_f_global})
df_f_global = df_f_global.sort_values(by='Valor F', ascending=False).reset_index(drop=True)

#Mostrar as Tabelas com os Valores F para Cada Cluster usando print:
for cluster in clusters_únicos:
    print(f"\nResultados para o Cluster {cluster}:\n")
    print(df_valores_f_dict[cluster])

#Mostrar a Tabela com os Valores F Globais usando print:
print("\nResultados Globais Considerando Todos os Clusters:\n")
print(df_f_global)

#Definir o Número Total de Gráficos Um para Cada Cluster + 1 Gráfico Global:
total_graficos = len(clusters_únicos) + 1

#Configurar o Layout dos Subplots Ex 3 Colunas:
n_colunas = 4
n_linhas = math.ceil(total_graficos / n_colunas)

figura, eixos = matplotlib.subplots(n_linhas, n_colunas, figsize=(n_colunas * 5, n_linhas * 6), constrained_layout=True)
eixos = eixos.flatten()

#Ajustar o Espaçamento entre os Gráficos:
figura.subplots_adjust(hspace=0.7, wspace=0.7)

#Gerar os Gráficos de Barras Individuais para Cada Cluster:
for i, cluster in enumerate(clusters_únicos):
    eixo = eixos[i]
    df_f = df_valores_f_dict[cluster]
    
    #Criar o Gráfico de Barras para o Cluster Atual:
    barras = eixo.bar(df_f['Variável'], df_f['Valor F'], color='#2ca02c', edgecolor='#2ca02c')
    
    #Adicionar o Valor F Acima de Cada Barra:
    for barra in barras:
        altura = barra.get_height()
        eixo.text(barra.get_x() + barra.get_width() / 2, altura, f'{altura:.2f}', 
                  ha='center', va='bottom', fontsize=8)
    
    eixo.set_title(f'Cluster {cluster}', fontsize=16)
    eixo.set_xlabel('Variável', fontsize=14)
    eixo.set_ylabel('Valor F', fontsize=14)
    eixo.tick_params(axis='x', rotation=45, labelsize=8)
    eixo.tick_params(axis='y', labelsize=8)
    
    #Ajustar o Limite Superior do Eixo Y:
    max_valor_f = np.max(df_f['Valor F'])
    
    # Calcular o Intervalo de Valores com base no Valor Máximo:
    intervalo = max(50, max_valor_f * 0.2)
    eixo.set_ylim(0, max_valor_f + intervalo)

#Gerar o Gráfico de Barras Global:
eixo_global = eixos[len(clusters_únicos)]
barras = eixo_global.bar(df_f_global['Variável'], df_f_global['Valor F'], color='#2ca02c', edgecolor='#2ca02c')
for barra in barras:
    altura = barra.get_height()
    eixo_global.text(barra.get_x() + barra.get_width() / 2, altura, f'{altura:.2f}', 
                     ha='center', va='bottom', fontsize=8)
eixo_global.set_title('Global', fontsize=16)
eixo_global.set_xlabel('Variável\n', fontsize=14)
eixo_global.set_ylabel('Valor F', fontsize=14)
eixo_global.tick_params(axis='x', rotation=45, labelsize=8)
eixo_global.tick_params(axis='y', labelsize=8)

#Ajustar o Limite Superior do Eixo Y para o Gráfico Global:
max_valor_f_global = np.max(df_f_global['Valor F'])
intervalo_global = max(50, max_valor_f_global * 0.2)
eixo_global.set_ylim(0, max_valor_f_global + intervalo_global)

#Ocultar Subplots Vazios:
for j in range(total_graficos, len(eixos)):
    eixos[j].set_visible(False)

#Definir o Título Geral da Figura:
figura.suptitle('Conjunto de Dados 1 - Algoritmo de Clustering Hierárquico Aglomerativo\n\nImportância das Variáveis para a Separação dos 12 Clusters com base no Valor F\n', fontsize=20)

#Mostrar o Gráfico:
matplotlib.show()
```

<H5>4.2. Análise Multivariada Não-Supervisionada de Valores Atípicos:</H5>

<H5>4.2.1. Modelo de Deteção de Valores Atípicos Local Outlier Factor (LOF):</H5>

<H5>4.2.1.1. Otimização dos Hiperparâmetros do Modelo de Deteção de Valores Atípicos Local Outlier Factor para Deteção de Valores Atípicos:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Ler a Biblioteca Numerical Python:
import numpy as np

#A partir do Submódulo Neighbors da Biblioteca Scikit-Learn, Importar o Modelo Local Outlier Factor:
from sklearn.neighbors import LocalOutlierFactor

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_deteção_valores_atípicos.csv')

#Definir os Valores do Hiperparâmetro Número de Vizinhos a Testar:
valores_hiperparâmetro_número_vizinhos = np.arange(15, 60, 5)

#Definir o Hiperparâmetro Medidas de Distância a Testar:
hiperparametro_medidas_distancia = ['euclidean', 'manhattan', 'cosine']

#Dicionário para Traduzir os Nomes das Medidas de Distância:
dicionário_medidas_distância = {'euclidean': 'Euclidiana', 'manhattan': 'Manhattan', 'cosine': 'Cosseno'}

#Criar uma Lista para Armazenar Todas as Combinações de Hiperparâmetros e suas Respetivas Pontuações dos Valores Atípicos:
lista_combinacões_hiperparâmetros_pontuações_valores_atípicos = []

#Iniciar a Melhor Combinação de Hiperparâmetros:
melhor_combinação_hiperparâmetros = None

#Iniciar a Melhor Pontuação dos Valores Atípicos do Modelo Local Outlier Factor:
melhor_pontuação_valores_atípicos_modelo_lof = np.inf

#Iterar sobre os Valores do Hiperparâmetro Número de Vizinhos:
for valor_hiperparâmetro_número_vizinhos in valores_hiperparâmetro_número_vizinhos:
    
    #Iterar sobre o Hiperparâmetro Medida de Distância:
    for hiperparâmetro_medida_distância in hiperparametro_medidas_distancia:
        
        #Iniciar o Modelo Local Outlier Factor:
        modelo_local_outlier_factor = LocalOutlierFactor(n_neighbors=valor_hiperparâmetro_número_vizinhos, contamination='auto', metric=hiperparâmetro_medida_distância)
        
        #Ajustar o Modelo Local Outlier Factor e Calcular as Previsões:
        Y_previsto = modelo_local_outlier_factor.fit_predict(conjunto_dados)

        #Calcular a Pontuação Média dos Valores Atípicos do Modelo Local Outlier Factor:
        pontuação_valores_atípicos_média_lof = -modelo_local_outlier_factor.negative_outlier_factor_.mean()

        #Armazenar na Lista os Resultados da Combinação Atual e respetivas Pontuações dos Valores Atípicos:
        lista_combinacões_hiperparâmetros_pontuações_valores_atípicos.append({
            'n Vizinhos': valor_hiperparâmetro_número_vizinhos,
            'Medida de Distância': dicionário_medidas_distância[hiperparâmetro_medida_distância],
            'Pontuação dos Valores Atípicos': f"{pontuação_valores_atípicos_média_lof:.3f}"})

        #Mostrar a Combinação Atual de Hiperparâmetros e a Pontuação dos Valores Atípicos:
        print(f"Número de Vizinhos: {valor_hiperparâmetro_número_vizinhos}, Medida de Distância: {dicionário_medidas_distância[hiperparâmetro_medida_distância]}, Pontuação dos Valores Atípicos: {pontuação_valores_atípicos_média_lof:.3f}")

        #Se a Pontuação Média de Valores Atípicos do Modelo Local Outlier Factor For Melhor do que a Pontuação dos Valores Atípicos Atual:
        if pontuação_valores_atípicos_média_lof < melhor_pontuação_valores_atípicos_modelo_lof:
            
            #Atualizar a Melhor Pontuação dos Valores Atípicos do Modelo Local Outlier Factor:
            melhor_pontuação_valores_atípicos_modelo_lof = pontuação_valores_atípicos_média_lof
            
            #Armazenar a Melhor Combinação de Hiperparâmetros:
            melhor_combinação_hiperparâmetros = {'Número de Vizinhos': valor_hiperparâmetro_número_vizinhos, 'Medida de Distância': dicionário_medidas_distância[hiperparâmetro_medida_distância]}

#Mostrar a Melhor Combinação de Hiperparâmetros para a Melhor Pontuação dos Valores Atípicos:
print(f"\nMelhor Combinação de Hiperparâmetros: {melhor_combinação_hiperparâmetros}, Melhor Pontuação dos Valores Atípicos Local Outlier Factor: {melhor_pontuação_valores_atípicos_modelo_lof:.3f}")

#Converter os Resultados de Todas as Combinações de Hiperparâmetros e suas respetivas Pontuações dos Valores Atípicos num DataFrame:
resultados_combinações_hiperparâmetros_pontuações_valores_atípicos = pd.DataFrame(lista_combinacões_hiperparâmetros_pontuações_valores_atípicos)

#Guardar os Resultados de Todas as Combinações de Hiperparâmetros e suas respetivas Pontuações dos Valores Atípicos:
resultados_combinações_hiperparâmetros_pontuações_valores_atípicos.to_csv('análise_valores_atipicos_lof_resultados_combinação_hiperparâmetros_pontuações_anomalia.csv', index=False)
print("\nOs Resultados de Todas as Combinações de Hiperparâmetros e suas respetivas Pontuações dos Valores Atípicos Foram Guardados com Sucesso em análise_valores_atipicos_lof_resultados_combinação_hiperparâmetros_pontuações_anomalia.csv")
```

<H5>4.2.1.2.1. Gráficos de Dispersão Bivariados para Deteção de Valores Atípicos nas Variáveis Preco_Bilhete e Percentagem_Ocupacao_Sala Utilizando o Modelo Local Outlier Factor:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Importar a Biblioteca Matplotlib:
import matplotlib.pyplot as plt

#A partir do Submódulo Ticker da Biblioteca Matplotlib, Importar o Formatador:
from matplotlib.ticker import FuncFormatter

#A partir do Submódulo Neighbors da Biblioteca Scikit-Learn, Importar o Modelo Local Outlier Factor:
from sklearn.neighbors import LocalOutlierFactor

#Definir o Tipo de Letra dos Textos (sans-serif):
plt.rcParams['font.family'] = 'sans-serif'

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_deteção_valores_atípicos.csv')

#Definir a Função para Inserir as Variáveis X e Y como Variáveis de Entrada para o Gráfico Bivariado:
def função_inserir_variáveis_X_Y(variável_X, variável_Y):
    
    #Obter as Variáveis de Entrada:
    variáveis_entrada = conjunto_dados[[variável_X, variável_Y]]

    #Definir o Valor do Hiperparâmetro n Vizinhos:
    valor_hiperparâmetro_n_vizinhos = 25

    #Definir o Hiperparâmetro Medida de Distância:
    hiperparâmetro_medida_distância = 'cosine'

    #Iniciar o Modelo Local Outlier Factor com a Melhor Combinação de Hiperparâmetros:
    modelo_local_outlier_factor = LocalOutlierFactor(n_neighbors=valor_hiperparâmetro_n_vizinhos, contamination='auto', metric=hiperparâmetro_medida_distância)

    #Criar os Rótulos com o Modelo Local Outlier Factor para Identificação de Valores Atípicos:
    rótulos_modelo_local_outlier_factor = modelo_local_outlier_factor.fit_predict(variáveis_entrada)

    #Criar a Máscara para Identificação de Valores Normais com Base nos Rótulos do Modelo Local Outlier Factor:
    máscara_valores_normais = [True if l == 1 else False for l in rótulos_modelo_local_outlier_factor]

    #Criar a Máscara para Identificação de Valores Atípicos com Base nos Rótulos do Modelo Local Outlier Factor:
    máscara_valores_atípicos = [True if l == -1 else False for l in rótulos_modelo_local_outlier_factor]

    #Obter o Número Total de Registos:
    número_total_registos = len(variáveis_entrada)

    #Calcular o Número de Valores Atípicos Identificados nas Variáveis em Análise:
    número_valores_atípicos_identificados_variáveis_análise = sum(máscara_valores_atípicos)

    #Calcular a Percentagem de Valores Atípicos Identificados nas Variáveis em Análise:
    percentagem_valores_atípicos_identificados_variáveis_análise = (número_valores_atípicos_identificados_variáveis_análise / número_total_registos) * 100

    #Calcular o Número de Valores Normais:
    número_valores_normais = número_total_registos - número_valores_atípicos_identificados_variáveis_análise

    #Mostrar o Número Total de Registos do Conjunto de Dados:
    print(f"Número Total de Registos do Conjunto de Dados: {número_total_registos:,}".replace(",", ".") + "\n")

    #Mostrar o Número de Valores Atípicos Identificados nas Variáveis em Análise:
    print(f"Número de Valores Atípicos Identificados nas Variáveis em Análise: {número_valores_atípicos_identificados_variáveis_análise:,}".replace(",", ".") + "\n")

    #Mostrar a Percentagem de Valores Atípicos Identificados nas Variáveis em Análise:
    print(f"Percentagem de Valores Atípicos Identificados nas Variáveis em Análise: {percentagem_valores_atípicos_identificados_variáveis_análise:.0f}%\n")

    #Configurações do Gráfico de Dispersão Bivariado - Figura - Tamanho:
    plt.figure(figsize=(10, 5))

    #Mostrar os Valores Normais no Gráfico de Dispersão - Configurações dos Valores Normais - Designação, Cores dos Pontos e das Bordas e Tamanho:
    plt.scatter(conjunto_dados[variável_X][máscara_valores_normais], conjunto_dados[variável_Y][máscara_valores_normais], label='Valores Normais', c='#2ca02c', edgecolor='black', s=20)

    #Mostrar os Valores Atípicos no Gráfico de Dispersão - Configurações dos Valores Normais - Designação, Cores dos Pontos e das Bordas e Tamanho:
    plt.scatter(conjunto_dados[variável_X][máscara_valores_atípicos], conjunto_dados[variável_Y][máscara_valores_atípicos], label='Valores Atípicos', c='#ff7f0e', edgecolor='black', s=20)

    #Configurações do Gráfico de Dispersão Bivariado - Título Geral - Designação, Cor e Tamanho:
    plt.title(f'Conjunto de Dados 1 - Modelo Local Outlier Factor\n\nGráfico de Dispersão Bivariado das Variáveis {variável_X} e {variável_Y}\npara Deteção de Valores Atípicos\n', color='black', fontsize=16, fontfamily='sans-serif')

    #Adicionar as Linhas de Grelha Verticais e Horizontais:
    plt.grid(True, alpha=0.5)

    #Definir o Formatador de Valores dos Eixos:
    formatador_valores_eixos = FuncFormatter(lambda x, pos: f'{x:,.0f}'.replace(',', '.'))

    #Configurações do Eixo X - Rótulo - Cor e Tamanho:
    plt.xlabel(variável_X, color='black', fontsize=14)

    #Configurações do Eixo X - Valores - Tamanho:
    plt.tick_params(axis='x', labelsize=14)

    #Configurações do Eixo X - Valores - Formatação:
    plt.gca().xaxis.set_major_formatter(formatador_valores_eixos)

    #Configurações do Eixo Y - Rótulo - Cor e Tamanho:
    plt.ylabel(variável_Y, color='black',  fontsize=14)

    #Configurações do Eixo Y - Valores - Tamanho:
    plt.tick_params(axis='y', labelsize=14)

    #Configurações do Eixo Y - Valores - Formatação:
    plt.gca().yaxis.set_major_formatter(formatador_valores_eixos)

    #Configurações da Legenda de Cores - Posição:
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

    #Obter o Eixo Atual:
    eixo_atual = plt.gca()

    #Configurações das Margens dos Eixos X e Y - Cor e Espessura:
    for margem in ['bottom', 'left']:
        eixo_atual.spines[margem].set_color('black')
        eixo_atual.spines[margem].set_linewidth(2.5)

    #Ocultar as Margens Superior e Direita:
    eixo_atual.spines['top'].set_visible(False)
    eixo_atual.spines['right'].set_visible(False)

    #Mostrar o Gráfico de Dispersão Bivariado:
    plt.show()

#Definir a Variável do Eixo X:
variável_X = 'Preco_Bilhete'

#Definir a Variável do Eixo Y:
variável_Y = 'Percentagem_Ocupacao_Sala'

#Chamar a Função para Inserir as Variáveis X e Y como Variáveis de Entrada para o Gráfico Bivariado:
função_inserir_variáveis_X_Y(variável_X, variável_Y)
```

```python
#Importar a Biblioteca Pandas: 
import pandas as pd

#A partir do Submódulo Neighbors da Biblioteca Scikit-Learn, Importar o Modelo Local Outlier Factor:
from sklearn.neighbors import LocalOutlierFactor

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('conjunto_dados_deteção_valores_atípicos.csv')

#Definir o Melhor Valor para o Hiperparâmetro Número de Vizinhos:
valor_hiperparametro_n_vizinhos = 25

#Definir o Hiperparâmetro Medida de Distância:
hiperparametro_medida_distancia = 'cosine'

#Iniciar o Modelo Local Outlier Factor:
modelo_local_outlier_factor = LocalOutlierFactor(n_neighbors=valor_hiperparametro_n_vizinhos, metric=hiperparametro_medida_distancia)

#Ajustar o Modelo Local Outlier Factor às Variáveis de Entrada:
modelo_local_outlier_factor.fit(conjunto_dados)

#Calcular as Pontuações dos Valores Atípicos:
pontuações_valores_atípicos = modelo_local_outlier_factor.negative_outlier_factor_

#Mostrar as Pontuações dos Valores Atípicos:
print("Pontuações dos Valores Atípicos:\n")
print(pontuações_valores_atípicos)

#Converter os Resultados das Pontuações dos Valores Atípicos num DataFrame:
resultados_pontuações_valores_atípicos = pd.DataFrame(pontuações_valores_atípicos, columns=['Pontuacao_Valores_Atipicos'])

#Guardar os Resultados das Pontuações dos Valores Atípicos:
resultados_pontuações_valores_atípicos.to_csv('análise_valores_atípicos_lof_resultados_pontuações_valores_atípicos.csv', index=False)
print("\n\nOs Resultados das Pontuações dos Valores Atípicos Foram Guardados com Sucesso em análise_valores_atípicos_lof_resultados_pontuações_valores_atípicos.csv")
```

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('análise_valores_atípicos_lof_resultados_pontuações_valores_atípicos.csv')

#Calcular as Medidas Estatísticas Descritivas para as Pontuações dos Valores Atípicos:
medidas_estatísticas_descritivas_pontuações_valores_atípicos = conjunto_dados.describe()

#Calcular a Moda das Pontuações dos Valores Atípicos:
moda_pontuações_valores_atípicos = conjunto_dados.mode().iloc[0]

#Adicionar a Moda às Medidas Estatísticas Descritivas das Pontuações dos Valores Atípicos:
medidas_estatísticas_descritivas_pontuações_valores_atípicos.loc['Moda'] = moda_pontuações_valores_atípicos

#Calcular a Amplitude das Pontuações dos Valores Atípicos:
amplitude_pontuações_valores_atípicos = conjunto_dados.max() - conjunto_dados.min()

#Adicionar a Amplitude às Medidas Estatísticas Descritivas das Pontuações dos Valores Atípicos:
medidas_estatísticas_descritivas_pontuações_valores_atípicos.loc['Amplitude'] = amplitude_pontuações_valores_atípicos

#Calcular a Variância das Pontuações dos Valores Atípicos:
variância_pontuações_valores_atípicos = conjunto_dados.var()

#Adicionar a Variância às Medidas Estatísticas Descritivas das Pontuações dos Valores Atípicos:
medidas_estatísticas_descritivas_pontuações_valores_atípicos.loc['Variância'] = variância_pontuações_valores_atípicos

#Traduzir os Índices das Medidas Estatísticas Descritivas de Inglês para Português:
medidas_estatísticas_descritivas_pontuações_valores_atípicos.rename(index={
    'count': 'Contagem',
    'mean': 'Média',
    'std': 'Desvio Padrão',
    'min': 'Mínimo',
    '25%': '1.º Quartil',
    '50%': 'Mediana',
    '75%': '3.º Quartil',
    'max': 'Máximo',
    'Amplitude': 'Amplitude',
    'Variância': 'Variância'}, inplace=True)

#Ordenar as Medidas Estatísticas Descritivas das Pontuações dos Valores Atípicos:
medidas_estatísticas_descritivas_pontuações_valores_atípicos_ordenadas = ['Contagem', 'Média', 'Mediana', 'Moda', 'Desvio Padrão', 'Mínimo', 
    '1.º Quartil', '3.º Quartil', 'Máximo', 'Amplitude', 'Variância']

#Reordenar as Medidas Estatísticas Descritivas das Pontuações dos Valores Atípicos segundo a Ordem Definida:
medidas_estatísticas_descritivas_pontuações_valores_atípicos = medidas_estatísticas_descritivas_pontuações_valores_atípicos.loc[medidas_estatísticas_descritivas_pontuações_valores_atípicos_ordenadas].round(2)

#Mostrar as Medidas Estatísticas Descritivas das Pontuações dos Valores Atípicos:
print(medidas_estatísticas_descritivas_pontuações_valores_atípicos)

#Guardar as Medidas Estatísticas Descritivas das Pontuações dos Valores Atípicos:
medidas_estatísticas_descritivas_pontuações_valores_atípicos.to_csv('análise_valores_atípicos_lof_resultados_medidas_estatísticas_descritivas.csv', index=False)
print("\n\nOs Resultados das Pontuações dos Valores Atípicos Foram Guardados com Sucesso em análise_valores_atípicos_lof_resultados_medidas_estatísticas_descritivas.csv")
```

<H5>4.2.2.5. Cálculo da Quantidade e da respetiva Percentagem de Valores Atípicos com Base nas Pontuações dos Valores Atípicos:</H5>

```python
#Importar a Biblioteca Pandas:
import pandas as pd

#Ler o Conjunto de Dados:
conjunto_dados = pd.read_csv('análise_valores_atípicos_lof_resultados_pontuações_valores_atípicos.csv')

#Definir o Fator do Intervalo Interquartil:
fator_intervalo_interquartil = 1.5

#Selecionar a Coluna de Pontuação dos Valores Atípicos:
coluna_pontuação_valores_atípicos = 'Pontuacao_Valores_Atipicos'

#Calcular o 1.º Quartil:
Q1 = conjunto_dados[coluna_pontuação_valores_atípicos].quantile(0.25)
    
#Calcular o 3.º Quartil:
Q3 = conjunto_dados[coluna_pontuação_valores_atípicos].quantile(0.75)
    
#Calcular o Intervalo Interquartil:
intervalo_interquartil = Q3 - Q1
    
#Calcular o Limite Inferior - Intervalo Interquartil:
limite_inferior = Q1 - fator_intervalo_interquartil * intervalo_interquartil
    
#Identificar os Valores Atípicos:
valores_atípicos = conjunto_dados[coluna_pontuação_valores_atípicos] < limite_inferior

#Calcular a Quantidade de Valores Atípicos:
quantidade_valores_atípicos = valores_atípicos.sum()
    
#Mostrar a Quantidade de Valores Atípicos:
print(f"Quantidade de Valores Atípicos: {quantidade_valores_atípicos:,.0f}".replace(",", "."))

#Calcular a Percentagem de Valores Atípicos:
percentagem_valores_atípicos = valores_atípicos.mean() * 100

#Mostrar a Percentagem de Valores Atípicos:
print(f"\nPercentagem de Valores Atípicos: {percentagem_valores_atípicos:.2f}%")
```
